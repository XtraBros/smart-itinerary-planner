```python
import pandas as pd

# Example DataFrame
data = pd.read_csv("../zoo-info.csv")
df = pd.DataFrame(data)

# Combine longitude and latitude into a new 'location' column
df['location'] = df.apply(lambda row: [row['longitude'], row['latitude']], axis=1)

# Show the updated DataFrame
```
